# Positive Integer Solutions of a Linear Equation

## Problem

Find the number of **positive integer solutions** of

$$
x+2y+3z=15,
$$

where

$$
x,y,z\in\mathbb{Z}^{+}.
$$

# Method 1: Fix One Variable and Count

From

$$
x+2y+3z=15,
$$

we have

$$
x=15-2y-3z.
$$

Since $x$ is a positive integer,

$$
15-2y-3z\geq 1.
$$

Therefore,

$$
2y+3z\leq 14.
$$

Since $y\geq 1$,

$$
2+3z\leq 14,
$$

so

$$
z\leq 4.
$$

Hence, the possible values of $z$ are

$$
z=1,2,3,4.
$$

# Method 2: Convert Positive Integers into Non-negative Integers

Since $x,y,z$ are positive integers, write

$$
x=a+1,\qquad y=b+1,\qquad z=c+1,
$$

where

$$
a,b,c\geq 0.
$$

Substituting into

$$
x+2y+3z=15,
$$

we get

$$
(a+1)+2(b+1)+3(c+1)=15.
$$

Therefore,

$$
a+2b+3c+6=15,
$$

and hence,

$$
a+2b+3c=9.
$$

## Counting the Non-negative Integer Solutions

For every fixed value of $c$,

$$
a+2b=9-3c.
$$

Since $a,b,c\geq 0$, the possible values of $c$ are

$$
c=0,1,2,3.
$$

For an equation of the form

$$
a+2b=N,
$$

we have

$$
a=N-2b.
$$

Therefore, $b$ may take the values

$$
0,1,2,\ldots,\left\lfloor\frac{N}{2}\right\rfloor.
$$

Hence, the number of non-negative integer solutions is

$$
\left\lfloor\frac{N}{2}\right\rfloor+1.
$$

Now consider each value of $c$.

### When $c=0$

$$
a+2b=9.
$$

Number of solutions:

$$
\left\lfloor\frac{9}{2}\right\rfloor+1=5.
$$

### When $c=1$

$$
a+2b=6.
$$

Number of solutions:

$$
\left\lfloor\frac{6}{2}\right\rfloor+1=4.
$$

### When $c=2$

$$
a+2b=3.
$$

Number of solutions:

$$
\left\lfloor\frac{3}{2}\right\rfloor+1=2.
$$

### When $c=3$

$$
a+2b=0.
$$

Number of solutions:

$$
1.
$$

Therefore,

$$
5+4+2+1=12.
$$

Hence,

$$
\boxed{12}
$$

# Method 3: Direct Floor-Function Formula

From

$$
x=15-2y-3z,
$$

and $x\geq 1$, we obtain

$$
2y+3z\leq 14.
$$

For a fixed positive integer $z$,

$$
2y\leq 14-3z.
$$

Therefore,

$$
y\leq \frac{14-3z}{2}.
$$

Since $y$ is a positive integer, the number of possible values of $y$ is

$$
\left\lfloor\frac{14-3z}{2}\right\rfloor.
$$

The largest possible value of $z$ is $4$. Hence,

$$
N=\sum_{z=1}^{4}
\left\lfloor\frac{14-3z}{2}\right\rfloor.
$$

Thus,

$$
N=
\left\lfloor\frac{11}{2}\right\rfloor
+
\left\lfloor\frac{8}{2}\right\rfloor
+
\left\lfloor\frac{5}{2}\right\rfloor
+
\left\lfloor\frac{2}{2}\right\rfloor.
$$

Therefore,

$$
N=5+4+2+1=12.
$$

Hence,

$$
\boxed{12}
$$

# Method 4: Generating Functions

For the positive integer variable $x$, the possible contributions are

$$
1,2,3,\ldots
$$

Therefore, its generating function is

$$
t+t^2+t^3+\cdots
=
\frac{t}{1-t}.
$$

For the term $2y$, where $y$ is positive, the possible contributions are

$$
2,4,6,8,\ldots
$$

Therefore, its generating function is

$$
t^2+t^4+t^6+\cdots
=
\frac{t^2}{1-t^2}.
$$

For the term $3z$, the possible contributions are

$$
3,6,9,12,\ldots
$$

Therefore, its generating function is

$$
t^3+t^6+t^9+\cdots
=
\frac{t^3}{1-t^3}.
$$

The combined generating function is

$$
\left(\frac{t}{1-t}\right)
\left(\frac{t^2}{1-t^2}\right)
\left(\frac{t^3}{1-t^3}\right).
$$

Thus,

$$
G(t)=
\frac{t^6}
{(1-t)(1-t^2)(1-t^3)}.
$$

Every choice of one term from each factor produces a term of the form

$$
t^{x+2y+3z}.
$$

Therefore, the number of positive integer solutions of

$$
x+2y+3z=15
$$

is the coefficient of $t^{15}$ in $G(t)$.

Hence,

$$
[t^{15}]
\frac{t^6}
{(1-t)(1-t^2)(1-t^3)}.
$$

Equivalently, this is

$$
[t^9]
\frac{1}
{(1-t)(1-t^2)(1-t^3)}.
$$

The coefficient is

$$
12.
$$

Therefore,

$$
\boxed{12}
$$

# Why the Generating Function Works

Consider the product

$$
(t+t^2+t^3+\cdots)
(t^2+t^4+t^6+\cdots)
(t^3+t^6+t^9+\cdots).
$$

Selecting $t^x$ from the first factor represents choosing the value $x$.

Selecting $t^{2y}$ from the second factor represents choosing the value $y$.

Selecting $t^{3z}$ from the third factor represents choosing the value $z$.

Their product is

$$
t^x\cdot t^{2y}\cdot t^{3z}
=
t^{x+2y+3z}.
$$

Thus, every positive integer solution of

$$
x+2y+3z=15
$$

contributes one term $t^{15}$.

Therefore, the coefficient of $t^{15}$ is exactly the number of positive integer solutions.

# Complete List of Positive Integer Solutions

The positive integer solutions $(x,y,z)$ are

$$
\begin{aligned}
&(10,1,1),\ (8,2,1),\ (6,3,1),\ (4,4,1),\ (2,5,1),\\
&(7,1,2),\ (5,2,2),\ (3,3,2),\ (1,4,2),\\
&(4,1,3),\ (2,2,3),\\
&(1,1,4).
\end{aligned}
$$

Therefore, the total number of solutions is

$$
\boxed{12}.
$$

# Python code

```python
solutions = []

for x in range(1, 16):
    for y in range(1, 16):
        for z in range(1, 16):
            if x + 2*y + 3*z == 15:
                solutions.append((x, y, z))

for solution in solutions:
    print(solution)
print('Total number of solutions are : {}'.format(len(solutions)))
```
